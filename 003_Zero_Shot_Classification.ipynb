```python
# Load Modules
from transformers import pipeline
```

About BART: https://research.facebook.com/publications/bart-denoising-sequence-to-sequence-pre-training-for-natural-language-generation-translation-and-comprehension/

```python
classifier = pipeline("zero-shot-classification", model="StevenLimcorn/indo-roberta-indonli")
```

#**Perform Zero Shot Classification**

```python
candidate_labels = ['teknologi', 'lingkungan', 'ekonomi', 'pendidikan', 'hukum', 'politik']
```

```python
sequence_to_classify = "Kemajuan terbaru dalam energi terbarukan sangat mengesankan, menunjukkan kemajuan yang signifikan dalam teknologi energi surya dan angin."
results = classifier(sequence_to_classify, candidate_labels, multi_label=True)
```

```python
# Formatting the output
formatted_output = "Classification Results:\n"
for label, score in zip(results['labels'], results['scores']):
    formatted_output += f"- {label.capitalize()}: {score:.2f}\n"

print(formatted_output)
```

#**Try Zero Shot Classification on Your Dataset (Excel)**

```python
# Function to Perform Zero-shot Classification
def perform_classification(text, candidate_labels):
    result = classifier(text, candidate_labels)
    return result['labels'][0], result['scores'][0]  # Return the top label and its score
```

```python
from google.colab import files
import io
import pandas as pd


# Create an upload button for the user to upload an Excel file
uploaded = files.upload()

# Process uploaded file
if uploaded:
    for fn in uploaded.keys():
        print(f'User uploaded file "{fn}" with length {len(uploaded[fn])} bytes')

        # Read the uploaded Excel file into DataFrame
        excel_data = pd.read_excel(io.BytesIO(uploaded[fn]))

        # Perform Zero-shot Classification on each row and add the results to the DataFrame
        excel_data[['Top_Label', 'Score']] = excel_data.apply(
            lambda row: perform_classification(row['Text'], candidate_labels) if pd.notnull(row['Text']) else ('', ''),
            axis=1,
            result_type='expand'
        )
```
