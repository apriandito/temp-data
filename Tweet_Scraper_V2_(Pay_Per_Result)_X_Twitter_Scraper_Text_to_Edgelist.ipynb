```python
import pandas as pd
import re
from google.colab import files
```

```python
df = pd.read_csv("tweets.csv")
df
```

```python
df_tweets = df[["author/userName", "text"]]
df_tweets.head(10)
```

```python
# Fungsi untuk mengekstrak mentions
def extract_mentions(text):
        return re.findall(r'@(\w+)', text)

def create_edgelist(df):

    # Mengaplikasikan fungsi untuk mengekstrak mentions
    df['mentions'] = df['text'].apply(extract_mentions)

    # Membuat edge list
    edges = []
    for index, row in df.iterrows():
        source = row['author/userName']
        for mention in row['mentions']:
            edges.append((source, mention))

    # Mengonversi ke DataFrame
    edge_df = pd.DataFrame(edges, columns=['source', 'target'])

    return edge_df
```

```python
df_edgelist = create_edgelist(df)
df_edgelist
```

```python
df_edgelist.to_csv("edgelist.csv", index=False)

# Provide a link to download the Excel file
files.download("edgelist.csv")
```
