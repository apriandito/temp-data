```python
# Load Modules
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

About BERT: https://arxiv.org/abs/1810.04805

```python
# Download Pretrained Model
pretrained= "mdhugol/indonesia-bert-sentiment-classification"
```

```python
# Set Model and Tokenizer
model = AutoModelForSequenceClassification.from_pretrained(pretrained)
tokenizer = AutoTokenizer.from_pretrained(pretrained)

# Set Pipeline
sentiment_analysis = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Set Label
label_index = {'LABEL_0': 'positive', 'LABEL_1': 'neutral', 'LABEL_2': 'negative'}
```

#**Perform Sentiment Analysis**

```python
# Input Text
text = "anjing enak banget"
```

```python
# Sentiment Analysis
result = sentiment_analysis(text)
status = label_index[result[0]['label']]
score = result[0]['score']
print(f'Text: {text} | Label : {status} ({score * 100:.3f}%)')
```

```python
# Function to Perform Sentiment Analysis
def perform_sentiment_analysis(text):
    result = sentiment_analysis(text)
    sentiment = label_index[result[0]['label']]
    score = result[0]['score']
    return sentiment, score
```

#**Try Sentiment Analysis on Your Dataset (Excel)**


```python
from google.colab import files
import io
import pandas as pd

# Create an upload button for the user to upload an Excel file
uploaded = files.upload()

# Process uploaded file
if uploaded:
    for fn in uploaded.keys():
        print(f'User uploaded file "{fn}" with length {len(uploaded[fn])} bytes')
        # Read the uploaded Excel file into DataFrame
        excel_data = pd.read_excel(io.BytesIO(uploaded[fn]))

        # Perform Sentiment Analysis on each row and add the results to the DataFrame
        excel_data[['Sentiment', 'Score']] = excel_data.apply(
            lambda row: perform_sentiment_analysis(row['Text']) if pd.notnull(row['Text']) else ('', ''),
            axis=1,
            result_type='expand'
        )
```
